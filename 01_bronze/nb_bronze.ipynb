```python

data = spark.read.format("csv").option("header", "true").option("inferschema", "false").load("s3a://de-ecommerce-miniproject/raw/metadata.csv")
for d in data.collect():
    temp = spark.read.format('csv').option("header","true").option("inferschema","false").load(f"s3a://de-ecommerce-miniproject/raw/{d.file_name}.{d.extension}")
    temp.write.mode("overwrite").format("delta").saveAsTable(f"ecommerce.bronze.{d.table_name}")
```
